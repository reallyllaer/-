```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
kiwi.tokenize(df['content'][0])[0]
```

```python
kiwi.tokenize(df['content'][0])[0][1]
```

```python
def tokenizer(sentence):
    try:
        if not sentence:
            raise ValueError
        words = [token[0] for token in kiwi.tokenize(sentence) if token.tag == "NNG" or "NNP" or "VV" or "VA" or "XR"]
    except ValueError as e:
        print(e)

    return words
```

```python
tokenizer(df['content'][0])
```
